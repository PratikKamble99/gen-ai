### Exception handling


```python
try:
    a=b
except:
    print('b is not defined')
```

```python
# exception handling with Exception class
try:
    a=b
except NameError as ex:
    print('Error:',ex)
```

```python
# exception handling with Exception class
try:
    a=1/0
except ZeroDivisionError as ex:
    print('Error:',ex)
```

```python
# exception handling with Main Exception class
try:
    a=b
    a=1/0
except ZeroDivisionError as ex:
    print('Error:',ex)
except Exception as main_ex:
    print('Error:', main_ex)
```

```python

try:
    a=int(input('Enter number'))
    x=10/a
except ValueError as ex:
    print('Error:',ex)
except ZeroDivisionError as main_ex:
    print('Error:', main_ex)
```

```python
# try, except, else block
try:
    a=int(input('Enter number'))
    x=10/a
except ValueError as ex:
    print('Error:',ex)
except ZeroDivisionError as zero_ex:
    print('Error:', zero_ex)
except Exception as main_ex:
    print('Error:', main_ex)
else:
    # this will run if no any error occur
    print(f"result is {x}")
```

```python
# try, except, else, finally block
try:
    a=int(input('Enter number'))
    x=10/a
except ValueError as ex:
    print('Error:',ex)
except ZeroDivisionError as zero_ex:
    print('Error:', zero_ex)
except Exception as main_ex:
    print('Error:', main_ex)
else:
    # this will run if no any error occur
    print(f"result is {x}")
finally:
    print('This runs every time')
```

```python
# file handling and xception handling
# file = None  # define file beforehand
try:
    file=open('examplwe.txt', 'r')
    content=file.read()
    print(content)

except FileNotFoundError as ex:
    print('Error:',ex)

finally:
    if(file in locals() or not file.closed):
        file.close()
        print('file closed')
```
