### Wordnet lemmatization

Lemmatization is the process of reducing a word to its lemma, WHich is root word rather than stem, the output of stemming. 
Unlike stemming (which just chops off endings), lemmatization understands the meaning and context so it returns a valid word.

NLTK provides WordNetLemmatizer class which is thin wrapper around the wordnet corpus. 


```python
import nltk
nltk.download('wordnet')
```

```python
## Q&A,chatbots,text summarization
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
'''
POS- Noun-n
verb-v
adjective-a
adverb-r
'''
lemmatizer.lemmatize("going",pos='v')
```

```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

```python
for word in words:
    print(word+"---->"+lemmatizer.lemmatize(word,pos='v'))
```

```python
lemmatizer.lemmatize("fairly",pos='v'),lemmatizer.lemmatize("sportingly")
```
