### BAsic Tokenization

```python
corpus="""Hi, I am Pratik Kamble.
And I am learning GEN-AI course which is krish naik's course.
"""
```

```python
print(corpus)
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import sent_tokenize
# convet paragraphs into documents/sentenses directaly
# corpus --> documents
documents = sent_tokenize(corpus)

print(documents)
```

```python
from nltk.tokenize import word_tokenize
# documents --> words
for document in documents:
    # print(document)
    print(word_tokenize(document))
```

```python
from nltk.tokenize import word_tokenize
# convet paragraphs into words directaly

# corpus --> words
# This does not seperate caractors like apha-strpohis"'s" Eg. "'s" --> ['s] -> treated as single word
word_tokenize(corpus)
```

```python
from nltk.tokenize import wordpunct_tokenize
# This seperate caractors like apha-strpohis"'s" eg. "'s" --> [', s] -> treated as separate word

wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer

# In this tokenization except last fullstop it not consider as seperate word
# "name. knksdn." --> tokenization --> ["name." , "knksdn", "."]
tokenizer = TreebankWordTokenizer()

tokenizer.tokenize(corpus)
```
