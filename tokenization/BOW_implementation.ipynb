```python
import pandas as pd

# import os
# print(os.listdir('./data'))
messages = pd.read_csv('./data/large_spam_collection.csv',)
messages
```

```python
# Data cleaining

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

import re

ps = PorterStemmer()
# nltk.download('stopwords')
```

```python
# apply regex to remove non laters words -> Lower sentence -> Apply stop words -> Stemming -> append in corpus list
corpus = []

for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if word not in stopwords.words('english')]
    review = " ".join(review)
    corpus.append(review)

print(corpus)
```

```python
# Create Bag Of Words using sklearn.countVectorizer
%pip install scikit-learn
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# For binary BOW enable binary=True
vectorizer = CountVectorizer(max_features=2500, binary=True)
X = vectorizer.fit_transform(corpus).toarray()
# X
X.shape # (sentence_length * vocabulary_length)

X

```

```python
vectorizer.vocabulary_
```

```python
# Vectorization using n-gram
""" 
    ngram_range=(1, 1) - unigram
                (1, 2) - unigram, bigram
                (1, 3) - unigram, trigram
                (2, 2) - bigram

"""
vectorizer = CountVectorizer(max_features=2500, binary=True, ngram_range=(2,2))
X = vectorizer.fit_transform(corpus).toarray()
# X
X.shape # (sentence_length * vocabulary_length)

X
```
